```python
data_dir = "../data/"
```

```python
ent2id, id2ent = dict(), dict()
rel2id, id2rel = dict(), dict()
for line in open(data_dir + 'newentity2id.txt', 'r'):
    [ent, id] = line.split()
    ent2id[ent] = int(id)
    id2ent[int(id)]  = ent
for line in open(data_dir + 'relation2id.txt', 'r'):
    [rel, id] = line.split()
    rel2id[rel] = int(id)
    id2rel[int(id)] = rel
print("entities num:\t" + str(len(ent2id)))
print("relations num:\t" + str(len(rel2id)))
```

```python
import numpy as np
ent_vec = []
for line in open("../res/entity2vec.bern"):
    vec = [float(i) for i in line.split()]
    ent_vec.append(np.array(vec))
```

```python
class Counter(dict):
    def __missing__(self, key):
        return False
```

```python
is_common = Counter()
kg1, kg2 = [], []
for line in open("../data/common_entities.txt"):
    is_common[ent2id[line.split()[0]]] = True
for i in range(24902):
    if(is_common[i]):
        continue
    if(i < 14951):
        kg1.append(i)
    else:
        kg2.append(i)        
```

```python
occupied = Counter()
# distance2entities_pair = []
cnt = 0
rank = 0
from numpy import linalg as LA
hits10 = hits1 = 0
for i in kg1:
    distance = []
    for j in kg2:
        dis = LA.norm(ent_vec[i]-ent_vec[j], 1)
        if(dis > 50):
            assert(id2ent[i] + '$' != id2ent[j])
            continue
        distance.append((dis, i, j)) #change to distance2entities_pair afterward
    distance.sort()
    for k in range(10):
        (dis, i, j) = distance[k]
        if(id2ent[i] + '$' == id2ent[j]):
            hits10 += 1
    for k in range(1):
        (dis, i, j) = distance[k]
        if(id2ent[i] + '$' == id2ent[j]):
            hits1 += 1
    for k in range(len(distance)):
        (dis, i, j) = distance[k]
        if(id2ent[i] + '$' == id2ent[j]):
            rank += k + 1
    cnt += 1      
    print(float(hits10)/cnt, float(hits1)/cnt, float(rank)/cnt, cnt)        
```
